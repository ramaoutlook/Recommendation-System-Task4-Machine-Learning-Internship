# Collaborative Filtering Based Recommender System

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating']
```

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating'].shape
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
x[x].index
```

```python
x.shape
```

```python
good_users = x[x].index
```

```python
ratings_with_name[ratings_with_name['User-ID'].isin(good_users)]
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(good_users)]
```

```python
filtered_rating.groupby('Book-Title').count()['Book-Rating']
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>50
y[y]
```

```python
famous_books = y[y].index
```

```python
filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

### Import the NearestNeighbors model from scikit-learn

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(pt).shape
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores
```

```python
import numpy as np
```

```python
# Function to recommend books based on a given book title
def recommend(book_name):
  #index fetch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x:x[1], reverse=True)[1:6]
  data = []
  for i in similar_items:
    item=[]
    tempdf = books_df[books_df['Book-Title'] == pt.index[i[0]]]
    item.extend(list(tempdf.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(list(tempdf.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(tempdf.drop_duplicates('Book-Title')['Image-URL-M'].values))

    data.append(item)
  return data
```

```python
recommend('Brave New World')
```

```python
recommend('The Secret Garden')
```
